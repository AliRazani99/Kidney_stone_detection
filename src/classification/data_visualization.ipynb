```python
import matplotlib.pyplot as plt
import torch
from dataset import get_transforms , ClassificationDataset
```

```python
root_dir = r"E:\master3\Improvment\kidney_stone_detection\Dataset"
dataset = ClassificationDataset(root_dir, transform=get_transforms('val'))
```

```python
img_tensor , label = dataset[0]
```

```python
mean = torch.tensor([0.485, 0.456, 0.406]).view(3,1,1)
std  = torch.tensor([0.229, 0.224, 0.225]).view(3,1,1)
img_unnorm = img_tensor * std + mean
```

```python
img_np = img_unnorm.permute(1,2,0).numpy()
```

```python
plt.figure(figsize=(4,4))
plt.imshow(img_np)
plt.title(f"Class: {dataset.classes[label]} ({label})")
plt.axis("off")
plt.show()
```
